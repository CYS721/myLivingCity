```python
# See how many csv file in data folder
import os
import pandas as pd

data_path = './myLivingCity_dataset'
csv_files = [f for f in os.listdir(data_path) if f.endswith('.csv')]
print(f'Total {len(csv_files)} csv files in data folder')
```

```python
# Load all csv files and see basic information
dfs = []
for f in csv_files:
    df = pd.read_csv(os.path.join(data_path, f))
    print(f'File: {f}')
    print(df.info())
    print()
    dfs.append(df)
    
```
