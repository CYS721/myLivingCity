```python
import subprocess
import json

def generate_sentence(query):
    # Construct the POST data as a dictionary
    data = {
        "model": "llama3",
        "messages": [
            {
                "role": "user",
                "content": query
            }
        ],
        "stream": False
    }

    # Convert dictionary to JSON format
    json_data = json.dumps(data)

    # Construct the curl command
    curl_command = [
        'curl',
        'http://192.168.1.124:11434/api/chat',
        '-d', json_data,
        '-H', 'Content-Type: application/json'
    ]

    # Execute the curl command
    process = subprocess.Popen(curl_command, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    output, error = process.communicate()

    if process.returncode == 0:
        return output
    else:
        return f"Error: {error.decode('utf-8')}"
```

```python
# Test the function
prompt = "Given a sentence, analyze its tone as either 'Positive', 'Negative', or 'Neutral' and identify the top keywords. Make sure Format your output as follows: 0. [Tone] | 1. [Attitude(verb)] | 2. [Keyword1] | 3. [Keyword2] | 4. [Keyword3] | 5. [Keyword4] | 6. [Keyword5]\n Ensure that the keywords are relevant and capture the essence of the sentence. Do not explain or give any other content! Just seven tag! You only only only need to give me the tags. The sentence is: "

each_sentence = "You will never find Losers who line up and pay good money for this crap line up to do charity or community work."

query = prompt + each_sentence

response = generate_sentence(query)
print(response)
data = json.loads(response)
print(*data["message"]["content"].split('\n')[0:], sep='\n')
```

```python
new_comment_set = set(new_comment_words)

# Function to find the existing list with the most common words
def find_most_common(new_comment_set):
    max_count = 0
    max_index = 0
    most_common_list = None

    for index, row in comments_df.iterrows():
        word_lst= row['response'][2:-2].split(" | ")
        lst = []
        for item in word_lst:
            try:
                word = item.split('. ')[1]
                if "None" not in word and "none" not in word:
                    lst.append(word)
            except IndexError:  
                continue  
        # print(lst)
        current_set = set(lst)
        common_words = current_set.intersection(new_comment_set)
        common_count = len(common_words)

        if common_count > max_count:
            max_index = index
            max_count = common_count
            most_common_list = lst

    return most_common_list, max_count, max_index

# Get the list with the most common words and the count of common words
result, count, index = find_most_common(new_comment_set)
print("Your input: \n", new_comment)
print("\nComments most similar to your input: \n", comments_df.at[index, 'comment_text'])
print("\nMost common words:", result)
print("Number of common words:", count)
```
